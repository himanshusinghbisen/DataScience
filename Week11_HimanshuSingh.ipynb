* What is the count of people belonging to different age groups in the persons table?
* Which age group has the maximum number of people?
* How many people do not have a last name?
* How many people have more than one pet?
* How many pets have received treatment?
* How many pets have received treatment, and the type of pet is known?
* How many pets are from the city called east port?
* How many pets are from the city called east port, and who received treatment?

```python
sql ="SELECT count(*), age FROM persons GROUP BY age"
result = cursor.execute(sql)

df = pd.DataFrame(result)
print("First column is Count and second column is Age")
print(df[[0, 1]].to_string(index=False, header=False))


sql1 ="SELECT age, COUNT(*) AS person_count FROM persons GROUP BY age ORDER BY person_count DESC LIMIT 1; "
result1 = cursor.execute(sql1)

df1 = pd.DataFrame(result1)
print("Age Group with maximum number of people")
print(df1.to_string(index=False, header=False))

sql2="SELECT COUNT(*) FROM persons WHERE last_name IS NULL;"
result = cursor.execute(sql2)

df = pd.DataFrame(result)
print("People Without a Last Name")
print(df.to_string(index=False, header=False))

sql3="SELECT COUNT(id) FROM (SELECT id FROM persons AS T1 JOIN pets AS T2 ON T1.id = T2.owner_id GROUP BY id HAVING COUNT(T2.pet_name) > 1) AS people_with_multiple_pets;"



result = cursor.execute(sql3)


df = pd.DataFrame(result)

print(f"People With More Than One Pet")
print(df.to_string(index=False, header=False))

sql4="SELECT COUNT(*) FROM pets  WHERE treatment_done = 1;"

result = cursor.execute(sql4)

df = pd.DataFrame(result)
print("Number of pets for which treatment is done")
print(df.to_string(index=False, header=False))


sql5="SELECT COUNT(*) FROM pets WHERE treatment_done = 1 AND pet_type IS NOT NULL;"

result = cursor.execute(sql5)

df = pd.DataFrame(result)
print("Number of pets for which treatment is done and pet type is known")
print(df.to_string(index=False, header=False))

sql6="SELECT COUNT(T2.pet_name) FROM persons AS T1 JOIN pets AS T2 ON T1.id = T2.owner_id WHERE T1.city = 'east port';"
result = cursor.execute(sql6)

df = pd.DataFrame(result)
print("Number of pets from City of East Port")
print(df.to_string(index=False, header=False))


sql7="SELECT COUNT(T2.pet_name) FROM persons AS T1 JOIN pets AS T2 ON T1.id = T2.owner_id WHERE T1.city = 'east port' AND T2.treatment_done = 1;"
result = cursor.execute(sql7)

df = pd.DataFrame(result)
print("Number of pets from City of East Port and have received treatment")
print(df.to_string(index=False, header=False))


```
