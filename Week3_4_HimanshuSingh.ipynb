Activity 3.01: Generating Statistics from a CSV File

```python

import pandas as pd


# Define the path to your .csv file
file_path = 'Boston_housing.csv'

# Load the data into a pandas DataFrame
try:
    # Reading the file
    df = pd.read_csv(file_path)
    print("Data loaded successfully!")
    # Reading first 10 rows
    print(df.head(10)) # Print the first 10 rows of the DataFrame
except FileNotFoundError:
    print(f"Error: The file at {file_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")

```

```python
# Showing number of Row counts
print(f"Number of Rows Are {len(df)}");

#column to keep
columns_to_keep = ['CRIM','ZN','INDUS','RM','AGE','DIS','RAD','TAX','PTRATIO','PRICE']
df_new = df[columns_to_keep]

# Subset of the original dataframe
df_new
```

```python
# Getting last 7 records of new dataframe
df_new.tail(7)
# Histogram for all variables
df_new.hist()
```

```python
# Plotting histogram using loop
import matplotlib.pyplot as plt

for col in df_new.columns:
    plt.title("Graph for the column "+col, fontsize=10)
    plt.hist(df_new[col])
    plt.show()
```

```python
# Plot Scatter plot for Crime rate Vs Price

plt.scatter(df_new['CRIM'], df_new['PRICE'])


```

```python
# Plot log10(crime) versus price.
import numpy as np

df['LOG_CRIM'] = np.log10(df['CRIM'])

plt.scatter(df['LOG_CRIM'], df['PRICE'])
plt.title('Log CRIME vs. PRICE')
plt.xlabel('Log of CRIME')
plt.ylabel('Price')
```

```python
# Calculate some useful statistics, such as mean rooms per dwelling, median age, mean distances to five Boston employment centers, and the percentage of houses with a low price (< $20,000).

# Mean of Rooms
print(df_new['RM'].mean())

# Median of Age
print(df_new['AGE'].median())

#Distance to Five Boston employment centers
print(df_new['DIS'].mean())

# Percentage of houses with price and less $20,000
Number_of_house_less_than_20000= df_new['PRICE']<20
count_less_than_threshold = len(df[Number_of_house_less_than_20000])
#print(count_less_than_threshold)
print(count_less_than_threshold/len(df_new)*100)
```

Activity 4.01: Working with the Adult Income Dataset (UCI)

```python
import csv

# Define the path to your .csv file
file_path_new = 'adult_income_data.csv'
df_income = pd.read_csv(file_path_new, header=None)
df_income.head()

# Load the data into a pandas DataFrame
try:
    # Reading the file
    with open(file_path_new, 'r') as file:

        # Use csv.reader to parse the data
        csv_reader = csv.reader(file)
        # Iterate over each row (line)
        for i, row in enumerate(csv_reader):
            if i == 0:
                print(f"Row {i}: {row}")


except FileNotFoundError:
    print(f"Error: The file at {file_path_new} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df_income.head()
```

```python


# list of column names
new_headers = ['age', 'workclass', 'fnlwgt','education','education-num','marital-status','occupation','relationship','sex','capital-gain','capital-loss','hours-per-week','native-country', 'Income']
# adding column names to dataframe

df_income.columns=new_headers
df_income.head()
```

```python
# Finding Missing values
missing_values_count = df_income.isnull().sum()
print(missing_values_count)
```

```python
# datafrane with Age, Education and occupation

df_income_subset = df_income[['age','education', 'occupation']]
df_income_subset.head()
```

```python
# Plot a histogram of age with a bin size of 20

df_income_subset['age'].hist(bins=20)
```

```python
# Create a function to strip the whitespace characters.

def strip_whitespace(text):
  """Removes leading and trailing whitespace from a string."""
  if isinstance(text, str):
    return text.strip()
  return text

# Use the apply method to apply this function to all the columns with string values, create a new column, copy the values from this new column to the old column, and drop the new column.

# Education column
df_income_subset['education_updated']= df_income_subset['education'].apply(strip_whitespace)
df_income_subset['education']=df_income_subset['education_updated']
df_income_subset.drop(['education_updated'], axis=1, inplace=True)
#df_income_subset

# Occupation column
df_income_subset['occupation_updated']= df_income_subset['occupation'].apply(strip_whitespace)
df_income_subset['occupation']=df_income_subset['occupation_updated']
df_income_subset.drop(['occupation_updated'], axis=1, inplace=True)
df_income_subset


```

```python
# Find the number of people who are aged between 30 and 50.

df_income_subset_30_50 = df_income_subset[(df_income_subset['age']>=30) & (df_income_subset['age']<=50)]
len(df_income_subset_30_50)
```

```python
# Group the records based on age and education to find how the mean age is distributed.
import seaborn as sns
import matplotlib.pyplot as plt

#df = sns.load_dataset('titanic')

plt.figure(figsize=(16, 5))

# Plot the boxplot, showing the mean
sns.boxplot(data=df_income_subset, y='age', x='education' , showmeans=True,

            # Customizing the mean marker to be a red diamond
            meanprops={
                "marker": "D",             # Diamond marker shape
                "markerfacecolor": "red",  # Fill color
                "markeredgecolor": "red",  # Outline color
                "markersize": "8"          # Size
            })

plt.title('Seaborn Boxplot with Custom Mean Marker')
plt.show()

```

```python
# Group by occupation and show the summary statistics of age. Find which profession has the oldest workers on average and which profession has its largest share of the workforce above the 75th percentile.

df_summary =df_income_subset.groupby('occupation').describe()['age']

#print(df_summary)

oldest_profession = df_summary['mean'].idxmax()
print(oldest_profession)

above_75th = df_summary['75%'].idxmax()
print(above_75th)



```

```python
# Use subset and groupBy to find the outliers.

occupation_data=df_income_subset.groupby('occupation').describe()['age']
plt.figure(figsize=(15,8))
plt.barh(y=occupation_data.index, width=occupation_data['count'])
plt.yticks(fontsize=13)
plt.show()
```

```python
    # Merge the two DataFrames using common keys to drop duplicate values.


df_1 = df_income[['age','workclass','education']].sample(5)
df_1.head()

df_2 = df_income[['education','occupation']].sample(5)
df_2.head()


merged_df = pd.merge(
    left=df_1,
    right=df_2,
    on='education',  # The column(s) shared by both DataFrames
    how='inner'  # Common keys only, dropping rows with unmatched keys
)

merged_df
```

     b. Series 2 = -2.1, 3.6, -1.5, 4, 3.1

c. Add Series 1 and Series 2 together and print the results

d. Subtract Series 1 from Series 2 and print the results

```python
import pandas as pd
print("Note: Operations align by index. Missing indices result in NaN.")
# # a. Create Series 1
data1 = [7.3, -2.5, 3.4, 1.5]
index1 = ['a', 'c', 'd', 'e']
series1 = pd.Series(data1, index=index1)

# b. Create Series 2
data2 = [-2.1, 3.6, -1.5, 4, 3.1]
index2 = ['a', 'c', 'e', 'f', 'g']
series2 = pd.Series(data2, index=index2)

# c. Add Series 1 and Series 2
sum_series = series1 + series2

print("\n--- c. Series 1 + Series 2 (Arithmetic Addition) ---")
print(sum_series.to_markdown(numalign="left", stralign="left"))

# d. Subtract Series 1 from Series 2 (Series 2 - Series 1)
sub_series = series2 - series1

print("\n--- d. Series 2 - Series 1 (Arithmetic Subtraction) ---")
print(sub_series.to_markdown(numalign="left", stralign="left"))
```
